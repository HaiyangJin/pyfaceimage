```python
import pyfaceimage as fim
```

# Original images

```python
im1 = fim.image('multi/L1.png', read=True)
im2 = fim.image('multi/L2.png', read=True)
```

```python
im1.pil
```

```python
im2.pil
```

# Make composite faces
## Top target 

```python
# default (ali, top)
cf_ali_top = fim.mkcf(im1, im2)
cf_ali_top.pil
```

```python
# misaligned
cf_mis_top = fim.mkcf(im1, im2, misali=100)
cf_mis_top.pil
```

## Bottom target

```python
# align bottom (now it is the same as default)
cf_ali_bot = fim.mkcf(im1, im2, cueistop=False)
cf_ali_bot.pil
```

```python
cf_mis_bot = fim.mkcf(im1, im2, cueistop=False, misali=100)
cf_mis_bot.pil
```

## Top target with cues

```python
# show cues
cf_ali_top_c = fim.mkcf(im1, im2, showcue=True)
cf_ali_top_c.pil
```

```python
cf_mis_top_c = fim.mkcf(im1, im2, misali=100, showcue=True)
cf_mis_top_c.pil
```

## Bottom target (with cues)

```python
cf_ali_bot_c = fim.mkcf(im1, im2, cueistop=False, showcue=True)
cf_ali_bot_c.pil
```

```python
cf_mis_bot_c = fim.mkcf(im1, im2, cueistop=False, misali=100, showcue=True)
cf_mis_bot_c.pil
```

## Different settings for misalignment

```python
# integer is used as pixels
cf_mis1 = fim.mkcf(im1, im2, misali=100)
cf_mis1.pil
```

```python
# use decimal as "ratio"
cf_mis2 = fim.mkcf(im1, im2, misali=0.5)
cf_mis2.pil
```

```python
try: 
    cf_mis2 = fim.mkcf(im1, im2, misali=2.5)
except:
    print("An exception occurred as expected.")
```
